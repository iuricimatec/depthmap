```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# equacao: f(x) = 3x^2 + 0.4x - 6
# tolerancia tol
x0 = 1.5  # x0 chute inicial
tol_err = 1e-4 # tolerancia
max = 100 # maximo de iteracoes

# Creating vectors X and Y
x = np.linspace(-2, 2, 100)
y = 3*x ** 2 + 0.4*x - 6
 
fig = plt.figure(figsize = (10, 5))
# Create the plot
plt.plot(x, y)
 
# Show the plot
plt.show()
```

```python
def f(x):
  return 3*(x*x) + 0.4*x - 6

def df(x):
  return 6*x + 0.4
```

```python
# Bloco MPF

def g(x):  # exemplo de equqcao que diverge
  print(x)
  return ( 6 - (3*x**2) ) / 0.4
  
def dg(x): # derivada desnecessaria. somente para entender o metodo MPF
  return -15*x

def h(x): # exemplo de equacao qeu coverge
  print(x)
  return math.sqrt( ( (6  - 0.4*x) / 3) )



def mpf(x0, tol, max):
  x = x0
  for i in range(max):
    x_prev = x
    x = h(x)
    if abs(x - x_prev) < tol:
      return x
  return None


root = mpf(1.5, 1e-4, 100)
print(f'Raizes: +- {root}')
```

```python
def tangente(x0, tol, max):
  x = x0
  for i in range(max):
    x_prev = x
    x = x - f(x)/df(x)  # equacao para calculo do metodo da tangente
    print(f'Interacao: {i} : {x}')
    if abs(x - x_prev)/x < tol:
      return x
  return None
```

```python
root = tangente(4.5, 1e-6, 100)
print(f'Raizes: +- {root}')
```
